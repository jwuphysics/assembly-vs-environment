```python
def get_flyover_nodes(edge_index):
    node_counts = torch.zeros((edge_index.max() + 1, 2), dtype=int)
    for src, dst in edge_index.t():
        node_counts[src, 0] += 1  # Outgoing
        node_counts[dst, 1] += 1  # Incoming
    
    single_connection_nodes = torch.where((node_counts[:, 0] == 1) & (node_counts[:, 1] == 1))[0]
    return single_connection_nodes
```

```python
def prune_merger_tree(edge_index):

    # compute node degree (number of incoming/outgoing directed edges)
    in_degree = torch.bincount(edge_index[1], minlength=edge_index.shape[1] + 1)
    out_degree = torch.bincount(edge_index[0], minlength=edge_index.max().item() + 1)
    
    # identify merges, i.e. has more than one incoming edge
    mergers = torch.where(in_degree > 1)[0]

    # identify leaf and root nodes
    leaf_node_indices = torch.where(in_degree == 0)[0]
    root_node_index = torch.where(out_degree == 0)[0]
    
    # map each node to its successor
    node_to_successor = {source.item(): target.item() for source, target in edge_index.t()}
    
    # determine prunable nodes (one edge in, one edge out, and not about to merge)
    prunable_nodes = []
    for node, succ in node_to_successor.items():
        if (in_degree[node] == 1) and (out_degree[node] == 1) and (succ not in mergers):
            prunable_nodes.append(node)
    prunable_nodes = torch.tensor(prunable_nodes, dtype=torch.long)
    print(prunable_nodes.shape)
    

    ### BUG SOMEWHERE BELOW!!!!!
    ###

    # identify predecessors and successors for pruned nodes
    predecessors = {}
    successors = {}
    for node in prunable_nodes:
        pred = edge_index[0, edge_index[1] == node].item()
        succ = edge_index[1, edge_index[0] == node].item()
        predecessors[node.item()] = pred
        successors[node.item()] = succ

    # create new edges bypassing pruned nodes
    new_edges_list = [[predecessors[node.item()], successors[node.item()]] for node in prunable_nodes]
    print(len(new_edges_list))
    
    # remove edges connected to pruned nodes
    mask = torch.ones(edge_index.size(1), dtype=torch.bool)
    for node in prunable_nodes:
        mask &= (edge_index[0] != node) & (edge_index[1] != node)
    filtered_edge_index = edge_index[:, mask]
    
    # combine old edges (excluding removed ones) with new edges
    if new_edges_list:
        new_edges = torch.tensor(new_edges_list).t().contiguous()
        pruned_edge_index = torch.cat([filtered_edge_index, new_edges], dim=1)
    else:
        pruned_edge_index = filtered_edge_index
    
    return pruned_edge_index
```
